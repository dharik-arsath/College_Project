## The Motive of this Project is to see the stores which sells their products at a lesser rate than other Stores

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/home/arsath/Store Price details.csv',index_col='Name of Store')
```

```python
data.rename(columns=lambda x:x.title(),inplace = True)
```

## Different Stores have different prices for different products,so the right way of handling this data is to see the overall price of all products in a Store (Total_price) so that we can go there and buy products in a cheaper rate.

```python
fig_size   = (16,6)
title_size = {'fontsize':20}
label_size = 20
font_size  = 20
```

```python
plt.figure(figsize=fig_size)
plt.title('Total Price in All Shops',fontdict=title_size)
sns.barplot(x = data.index,y = 'Total_Price',data = data)
plt.xlabel('Name of the Shop',fontsize=font_size)
plt.ylabel('Total_Price',fontsize = font_size)
plt.tick_params(labelsize = label_size)
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

## The Above Plot shows that pretty much all the shops sells their product at a same rate with some minor variance in price

```python
plt.figure(figsize=fig_size)
plt.title('Total Price in All Shops',fontdict=title_size)
plt.plot(data.index,'Total_Price','--*',data = data)
plt.xlabel('Name of the Shop',fontsize = font_size)
plt.ylabel('Total Price in All Shops',fontsize = font_size)
plt.tick_params(labelsize = label_size)
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

## As per this data we should never go to Harish Supermarket.

```python
min_prices = data.min()
min_prices
```

```python
## Get the lowest price of all the products
```

```python
all_products = data.columns.drop('Total_Price')
```

```python
def clean_data(word):
    index = 0
    while index != len(word) and word[index] != '(':
        index += 1
        continue
    return word[0:index].strip()
    
available_products = [clean_data(x) for x in all_products]
```

```python
def get_user_input():

    print(available_products)
    print('\n')
    prod   = input('\n\nEnter the Name of the Product to get least price among Shops: ').title()
    result = all_products[all_products.str.contains(prod)].values
    
    if len(result) == 1:
        price     = min_prices[result]
        for x in price:
            price = x
        result    = ''.join(map(str,result)) 
        
        shop_name =  data[data[result] == price].index.values
        shop_name = ''.join(map(str, shop_name)) 
        print(f"\nThe Minimum Price of {result} is {price} in {shop_name} ")
        
    elif len(result) >= 2:
        #print(f'These are the results available under your search query {result}\n')
        prod = pd.Series(result,index = range(1,len(result)+1))
        print(f"{prod}\n")
        
        print('Please enter the product Id to find the least price for the product: ')
        prod_id = int(input())
        result = prod[prod_id]
        price = min_prices[result]
        
        shop_name = data[data[result] == price].index.values
        shop_name = ''.join(map(str,shop_name))
        
        
    plt.figure(figsize=fig_size)
    plt.title(f'Price of {result} in All Shops',fontdict=title_size)
    plt.plot(data.index,result,'--*',data = data)
    plt.xlabel('Shop Name',fontsize = font_size)
    plt.ylabel(f'Price of a {result} in ',fontsize = font_size)
    plt.tick_params(labelsize = label_size)
    plt.xticks(rotation = 90)
    plt.grid(True)
    plt.show()
    
    print(f"\nThe Minimum Price of {result} is {price} in {shop_name}")
```

```python
get_user_input()
```

### The Above plot shows that which shop gives least price for a particular product
